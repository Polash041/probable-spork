```python
graph ={
    '1':['3','5'],
    '3':['1','2'],
    '2':['3'],
    '5':['1'],
    
}
visited=[]
parent={}

def dfs(node):
    print(node,end=" ")
    visited.append(node)
    for child in graph[node]:
        if child not in visited:
            visited.append(child)
            parent[child]=node
            dfs(child)
            

def path(start,target):
    if start==target:
        print(start)
        return
    else:
        print(target,end="->")
        path(start,parent[target])
    
    
dfs('1')
print(parent)
path('1','5')
```

```python
graph = {
    'S': ['A', 'B'],
    'A': ['C', 'D'],
    'B': ['G', 'H'],
    'C': ['E', 'F'],
    'G': ['I'],
    'E': ['K'],
    'D': [], 
    'H': [],  
    'F': [], 
    'I': [], 
    'K': [],
}

visited=[]
queue=[]

def bfs(start):
   
    visited=[]
    queue=[]
    visited.append(start)
    queue.append(start)
    
    while queue:
        u=queue.pop(0)
        print(u,end=" ")
        
        for child in graph[u]:
            if child not in visited:
                visited.append(child)
                queue.append(child)
                


bfs('S')
```

```python
graph={
    'A':{'B':11,'D':7,'C':14},
    'B':{'A':11,'E':15},
    'E':{'H':15,'B':15},
    'F':{'C':10,'D':25,'G':20},
    'H':{'G':10,'E':9},
    'G':{'F':20,'H':10},
    'C':{'A':14,'F':10,'E':8},
    'D':{'A':7,'F':25}
    
}

hvalue={
    'A':40,
    'B':32,
    'C':25,
    'D':35,
    'E':19,
    'F':17,
    'H':10,
    'G':0
    
}

parent={}
def astar(start,goal):
    visited=set()
    priority_queue=[(hvalue[start],start)]
    
    while priority_queue:

        h,current=priority_queue.pop(0)
        print(priority_queue)
        if current == goal:
            return
        if current in visited:
            continue
            
        visited.add(current)
        
        for child,cost in graph.get(current,{}).items():
            if child not in visited:
                final=cost+hvalue[child]
                priority_queue.append((final,child))
                parent[child]=current
                
                

def path(start,goal):
    rasta=[goal]
    
    while goal in parent and goal != start:
        goal=parent[goal]
        rasta.insert(0,goal)
        
    if rasta[0]==start:
        print("->".join(rasta))
    else:
        print("No Path")
            



astar('A','G')
path('A','G')
```

```python

Max=1000
Min=-1000

def minimax(depth,ni,maxplayer,values,alpha,beta):
    if depth==3:
        return values[ni]
    
    if maxplayer:
        best=Min
        
        for i in range(0,2):
            val=minimax(depth+1,ni * 2  + i,False,values,alpha,beta)
            best=max(best,val)
            alpha=max(alpha,best)
            
            if alpha >= beta:
                break
                
        return best
    
    else:
        best=Max
        
        for i in range(0,2):
            
            val=minimax(depth+1,ni*2+i,True,values,alpha,beta)  
            best=min(best,val)
            beta=min(beta,best)
            
            if alpha >= beta:
                break
                
        return best
            




values=[3,5,6,9,1,2,0,-1]
print(minimax(0,0,True,values,Min,Max))
```
